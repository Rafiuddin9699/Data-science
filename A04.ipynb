```python
'''Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
A1. A class is a blue print for creating objects. it defines the prooperties and behaviour that all of object of that class will have.
On the other hand , the object is an instance of a class. It represents a particular entity or concept, and it encapsulates data and beehaviour.
'''
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0

    def get_descriptive_name(self):
        return f"{self.year} {self.make} {self.model}"

    def read_odometer(self):
        return f"This car has {self.odometer_reading} miles on it."

    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")

    def increment_odometer(self, miles):
        self.odometer_reading += miles

# Creating an instance of the Car class
my_car = Car("Toyota", "Camry", 2020)

# Accessing attributes and methods of the object
print(my_car.get_descriptive_name())
print(my_car.read_odometer())

# Modifying attributes using methods
my_car.update_odometer(15000)
print(my_car.read_odometer())

# Incrementing attribute value
my_car.increment_odometer(100)
print(my_car.read_odometer())

    
```

```python
Q2. Name the four pillars of OOPs.
A2. The four pillars of Object-Oriented Programming (OOP) are:

1. Encapsulation: Encapsulation is the bundling of data (attributes) and methods (behaviors) that operate on the data into a single unit, called a class. It helps in hiding the internal state of an object and only exposes the necessary functionalities to interact with it.

2. Abstraction: Abstraction is the concept of simplifying complex systems by hiding unnecessary details while showing only the essential features of an object. It allows us to focus on what an object does without needing to know how it achieves it.

3. Inheritance: Inheritance is the mechanism by which a class (subclass or derived class) inherits properties and behaviors (attributes and methods) from another class (superclass or base class). It promotes code reusability and helps in creating a hierarchy of classes.

4. Polymorphism: Polymorphism allows objects to be treated as instances of their superclass. It enables a single interface to represent multiple types. Polymorphism is achieved through method overriding (redefining a method in a subclass) and method overloading (defining multiple methods with the same name but different parameters).
```

```python
'''Q3. Explain why the __init__() function is used. Give a suitable example.
A3. This also known as constructor , it is a spexcial meyhod in python classes. it is automaticallu called when a new instance (object) of the class is crated . The primmary purpose is to intialise the attibutes of the newly creates object.
'''
class Square:
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

# Create an instance of the Square class
square = Square(5)

# Call the area() method to calculate the area of the square
area = square.area()

# Print the area of the square
print(area)
```

```python
'''Q5. What is inheritance? Give an example for each type of inheritance.
A5. Inheritance is a fundamental concept in object-oriented programming (OOP) where a class (subclass or derived class) inherits properties and behaviors (attributes and methods) from another class (superclass or base class). It promotes code reusability and allows the creation of a hierarchy of classes, where subclasses inherit and extend the functionality of their superclass.'''

Single Inheritance:
python

# Superclass
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Subclass inheriting from Animal
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Creating an instance of Dog
dog = Dog()
dog.sound()  # Inherits from Animal
dog.bark()   # Unique to Dog class



Multiple Inheritance:
python

# Superclasses
class Parent1:
    def method1(self):
        print("Method 1 from Parent 1")

class Parent2:
    def method2(self):
        print("Method 2 from Parent 2")

# Subclass inheriting from Parent1 and Parent2
class Child(Parent1, Parent2):
    def method3(self):
        print("Method 3 from Child")

# Creating an instance of Child
child = Child()
child.method1()  # Inherits from Parent1
child.method2()  # Inherits from Parent2
child.method3()  # Unique to Child class


Multilevel Inheritance:

# Superclass
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Subclasses inheriting from Animal
class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Puppy(Dog):
    def playful_bark(self):
        print("Puppy playfully barks")

# Creating an instance of Puppy
puppy = Puppy()
puppy.sound()        # Inherits from Animal
puppy.bark()         # Inherits from Dog
puppy.playful_bark() # Unique to Puppy class


Hierarchical Inheritance:

# Superclass
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Subclasses inheriting from Animal
class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Creating instances of Dog and Cat
dog = Dog()
cat = Cat()
dog.sound()  # Inherits from Animal (same as cat.sound())
dog.bark()   # Unique to Dog class
cat.meow()   # Unique to Cat class
```
